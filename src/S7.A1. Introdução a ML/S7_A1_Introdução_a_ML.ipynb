## Question 5: Describe Generative AI and NLP

**a. IA Generativa**
É uma IA que **cria conteúdo novo** (texto, imagens, áudio, código) aprendendo padrões de grandes volumes de dados. Exemplos: ChatGPT, DALL·E, Claude. Usada em chatbots, geração de imagens e descoberta de medicamentos.

**b. PLN (Processamento de Linguagem Natural)**
Permite que computadores **entendam e gerem linguagem humana**. Inclui tarefas como tradução, análise de sentimento e reconhecimento de voz. Exemplos: Google Tradutor, Siri, filtros de spam.

## Question 8: Essential Skills for Machine Learning Professionals

Matemática e Estatística é a base do ML, envolvendo álgebra linear, cálculo, probabilidade e testes de hipóteses, essenciais para entender e construir modelos.

Programação é a capacidade de escrever código eficiente, principalmente em Python, utilizando bibliotecas como TensorFlow, PyTorch e Scikit-learn para desenvolver soluções de ML.
Manipulação de Dados é a habilidade de coletar, limpar e transformar dados brutos em informações úteis, incluindo análise exploratória e engenharia de features.

Conhecimento em Algoritmos de ML é o domínio de técnicas como regressão, classificação, clustering e redes neurais, sabendo escolher o modelo certo para cada problema.

Avaliação de Modelos é a capacidade de medir o desempenho de um modelo usando métricas como F1-score e RMSE, aplicando validação cruzada para garantir resultados confiáveis.

Conhecimento de Domínio é a habilidade de entender o contexto do problema de negócio e traduzi-lo em uma solução de ML adequada, comunicando resultados a públicos não técnicos.

MLOps e Deploy é a competência de colocar modelos em produção usando ferramentas de cloud (AWS, GCP, Azure), Docker e pipelines de monitoramento contínuo.
